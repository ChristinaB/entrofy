```python
num_applicants = 30

# uses numpy.random.choice to weight each category
# _categories and _probability can be modified and extended, but each array must be of equal length
# contains missing values, because we expect those in source data

gender_categories = ['male','female','other','']
# gender_probability = [.48,.48,.03,.01]
gender_probability = [.75,.20,.03,.02]

career_categories = ['student','junior-faculty','senior-faculty','']
# career_probability = [.33,.33,.33,.01]
career_probability = [.10,.60,.29,.01]

country_categories = ['OECD','non-OECD',''] # see http://www.oecd.org/about/membersandpartners/
# country_probability = [.49,.49,.02]
country_probability = [.90,.08,.02]

subfield_categories = ['solar','cosmology','astrophysics','exoplanets','']
# subfield_probability = [.24,.24,.24,.24,.04]
subfield_probability = [.05,.60,.15,.15,.05]

first_time_categories = ['first_time','returning','']
# first_time_probability = [.49,.49,.02]
first_time_probability = [.33,.66,.01]
```

```python
gender_array = numpy.random.choice(gender_categories, p=gender_probability,size=num_applicants,replace=True)

name_array = [names.get_full_name(gender=i) for i in gender_array]

career_stage_array = numpy.random.choice(career_categories, p=career_probability,size=num_applicants,replace=True)

country_array = numpy.random.choice(country_categories, p=country_probability,size=num_applicants,replace=True)

subfield_array = numpy.random.choice(subfield_categories, p=subfield_probability,size=num_applicants,replace=True)

first_time_array = numpy.random.choice(first_time_categories,p=first_time_probability,size=num_applicants,replace=True)
```

```python
df2 = pd.DataFrame({
        'applicant_name':name_array,
        'gender':gender_array,
        'career_stage':career_stage_array,
        'country':country_array,
        'subfield':subfield_array,
        'first_time':first_time_array
        })

df2
```

```python
fname = 'datasets/all-uneven-%d.csv' % num_applicants
df2.to_csv(fname,index=False)
```
